```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# Generar datos de ejemplo
np.random.seed(0)
data = np.random.rand(100, 5)  # 100 muestras, 5 caractersticas

# Separar en conjuntos de entrenamiento y prueba
train_data, test_data = train_test_split(data, test_size=0.2, random_state=0)

# Realizar PCA en el conjunto de entrenamiento
pca = PCA(n_components=2)  # Reducir a 2 componentes principales
train_pca = pca.fit_transform(train_data)

# Aplicar la transformacin PCA al conjunto de prueba
test_pca = pca.transform(test_data)

# Realizar PCA en el conjunto completo de datos para comparacin
pca_full = PCA(n_components=2)
data_pca_full = pca_full.fit_transform(data)

# Visualizar los resultados
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(train_pca[:, 0], train_pca[:, 1], label='Train', alpha=0.7)
plt.scatter(test_pca[:, 0], test_pca[:, 1], label='Test', alpha=0.7)
plt.title('PCA en Conjuntos de Entrenamiento y Prueba')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(data_pca_full[:, 0], data_pca_full[:, 1], label='Full Data', alpha=0.7)
plt.title('PCA en el Conjunto Completo de Datos')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt

# Cargar el conjunto de datos de dgitos
digits = load_digits()
data = digits.data
labels = digits.target

# Separar en conjuntos de entrenamiento y prueba
train_data, test_data, train_labels, test_labels = train_test_split(data, labels, test_size=0.2, random_state=0)

# Realizar PCA en el conjunto de entrenamiento
pca = PCA(n_components=2)  # Reducir a 2 componentes principales
train_pca = pca.fit_transform(train_data)

# Aplicar la transformacin PCA al conjunto de prueba
test_pca = pca.transform(test_data)

# Realizar PCA en el conjunto completo de datos para comparacin
pca_full = PCA(n_components=2)
data_pca_full = pca_full.fit_transform(data)

# Comparar la varianza explicada
explained_variance_train = pca.explained_variance_ratio_
explained_variance_full = pca_full.explained_variance_ratio_

print("Varianza explicada por los componentes principales en el conjunto de entrenamiento:")
print(explained_variance_train)

print("Varianza explicada por los componentes principales en el conjunto completo:")
print(explained_variance_full)

# Visualizar los resultados
plt.figure(figsize=(18, 6))

# PCA en el conjunto de entrenamiento y prueba
plt.subplot(1, 3, 1)
scatter = plt.scatter(train_pca[:, 0], train_pca[:, 1], c=train_labels, cmap='viridis', alpha=0.7)
plt.colorbar(scatter, label='Label')
plt.title('PCA en Conjunto de Entrenamiento')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')

plt.subplot(1, 3, 2)
scatter = plt.scatter(test_pca[:, 0], test_pca[:, 1], c=test_labels, cmap='viridis', alpha=0.7)
plt.colorbar(scatter, label='Label')
plt.title('PCA en Conjunto de Prueba')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')

# PCA en el conjunto completo de datos
plt.subplot(1, 3, 3)
scatter = plt.scatter(data_pca_full[:, 0], data_pca_full[:, 1], c=labels, cmap='viridis', alpha=0.7)
plt.colorbar(scatter, label='Label')
plt.title('PCA en Conjunto Completo de Datos')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')

plt.tight_layout()
plt.show()
```
